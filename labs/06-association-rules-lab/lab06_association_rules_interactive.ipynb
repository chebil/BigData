## Part 5: Business Recommendations

```python
# TODO: Find rules with high lift (> 1.0)
strong_rules = rules[rules['lift'] > 1.0].sort_values('lift', ascending=False)

print('BUSINESS RECOMMENDATIONS')
print('=' * 80)

for idx, rule in strong_rules.head(5).iterrows():
    antecedent = ', '.join(list(rule['antecedents']))
    consequent = ', '.join(list(rule['consequents']))
    
    print(f'\nRule: If customer buys {antecedent} â¡ï¸ Recommend {consequent}')
    print(f'  Support: {rule["support"]:.2%}')
    print(f'  Confidence: {rule["confidence"]:.2%}')
    print(f'  Lift: {rule["lift"]:.2f}')
    
    # Business action
    if rule['lift'] > 1.2:
        print(f'  âœ… Strong association - Consider product bundling or cross-promotion')
    else:
        print(f'  ğŸ”µ Moderate association - Monitor and test')
```
